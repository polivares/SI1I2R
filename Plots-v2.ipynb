# SIR interactive model 

## Change B' and D'

```python
# Originals values
N = 100000
SIIR0 = np.zeros(9)
SIIR0[1] = 10
SIIR0[2] = 10
SIIR0[0] = N - np.sum(SIIR0[1:8])

t_start = 0
t_end = 50
n_int = 100000

t_sim = np.linspace(t_start, t_end, n_int)
```

```python
# SET GOOD VALUES! (MEMORIZE)
# find good parameters values: 
# Criteria:
# - Smaller final size (0.2 or 0.1 if possible)
# - Minimal parameter set in order to observe resonance and the oposite to clarify condition
# - We are trying to found condition from competition or cooperation

#params = [6, 5.5, 5, 5, 6, 5.5, 100, 5]  # Params
#params = [6, 4, 2, 2, 6, 4, 100, 2]  # Params
#params = [6, 5.5, 5, 5, 6, 5.5, 6/1.9, 5.5/1.9]  # Params
params = [6, 5.5, 5, 5.4, 20, 5.5, 5, 5.4]  # Params
#params = [2, 5.5, 2/1.4, 5, 2, 5.5, 6/1.4, 5.5/1.4]  # Params

fig, ax = plt.subplots(2, 2)

params_copy = params.copy()
beta_prime_arr = np.array([0.5, 1, 2.5, 5.5, 6, 8.5, 11.5, 14.5, 17.5])
delta_prime_arr = beta_prime_arr/1.2
f_size1 = []
f_size2 = []
for i in range(len(beta_prime_arr)):
    params_copy[1] = beta_prime_arr[i]
    params_copy[5] = beta_prime_arr[i]
    params_copy[3] = delta_prime_arr[i]
    params_copy[7] = delta_prime_arr[i]
    siirSim = mdl.SIIR(SIIR0, params_copy, t_sim)
    siirSim.runEvaluation(norm=True)
    res = siirSim.getResult()
    
    inf1 = res[:, 1] + res[:, 3] + res[:, 7]
    rec1 = res[:, 4] + res[:, 6] + res[:, 8]
    f_size1.append(rec1[-1])
    ax[0,0].plot(t_sim, inf1, label='Beta2=' + str(beta_prime_arr[i]))
    ax[0,1].plot(t_sim, rec1, label='Beta2=' + str(beta_prime_arr[i]))
    
    inf2 = res[:, 2] + res[:, 3] + res[:, 6]
    rec2 = res[:, 5] + res[:, 7] + res[:, 8]
    f_size2.append(rec2[-1])
    ax[1,0].plot(t_sim, inf2, label='Beta2=' + str(beta_prime_arr[i]))
    ax[1,1].plot(t_sim, rec2, label='Beta2=' + str(beta_prime_arr[i]))

ax[0,0].legend()
ax[0,0].set_title('Disease 1')
ax[0,0].set_xlabel('Time')
ax[0,0].set_ylabel('N infected')
ax[0,1].legend()
ax[0,1].set_title('Disease 1')
ax[0,1].set_xlabel('Time')
ax[0,1].set_ylabel('N recovered')
ax[1,0].legend()
ax[1,0].set_title('Disease 2')
ax[1,0].set_xlabel('Time')
ax[1,0].set_ylabel('N infected')
ax[1,1].legend()
ax[1,1].set_title('Disease 2')
ax[1,1].set_xlabel('Time')
ax[1,1].set_ylabel('N recovered')

txt = ("Beta1 = " + str(params[0]) + ", Delta1 = " + str(params[2]) +
       ", Beta2 = " + str(params[1]) + ", Delta2 = " + str(params[3]) +
       "\nBeta1' = " + str(params[4]) + ", Delta1' = " + str(params[6]) +
       ", Beta2' = " + str(params[5]) + ", Delta2' = " + str(params[7]))
plt.figtext(0.65, 0.01, txt, ha="right", va="bottom",
            bbox=dict(boxstyle='square', facecolor='white', alpha=1))
fig.tight_layout()
plt.show()
```
